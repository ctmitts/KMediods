```python
def subSet( counts, sparsity):
    
    marginals = counts['marginals']
    joints = counts['joints']
    tris = counts['triples']
     ## Control for sparsity
    words = {}
    for wrd, c in marginals.iteritems():
            if c >= sparsity:
                    words[wrd] = c
    pairs = {}
    for j, jC in joints.iteritems():
            [wrd1, wrd2] = j.split(":")
            if wrd1 in words and wrd2 in words:
                    pairs[j] = jC
    triples = {}
    for tri, triC in tris.iteritems():
            [pObj, cWord] = ing.splitWords(tri)
            [wrd1, wrd2] = pObj.split(" ");
            pair1 = ing.sort_alpha(wrd1, cWord); 
            pair2 = ing.sort_alpha(wrd2, cWord)
            if ( pair1 in pairs and wrd1 in words ) and ( pair2 in pairs and wrd2 in words ):
                    triples[tri] = triC
                    
    return {'marginals':words, 'joints':pairs, 'triples':triples }
```
